Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
 - Type I error is rejecting a true null hypothesis, while a Type II error is failing to reject a false null hypothesis.

 Type I error: Occurs when the null hypothesis is true, but it is mistakenly rejected. This is often referred to as a "false positive". It impacts decision-making by leading to unnecessary changes or actions based on a false premise.
 Type II error: Occurs when the null hypothesis is false, but it is mistakenly not rejected. This is often referred to as a "false negative". It impacts decision-making by preventing necessary changes or actions, leading to missed opportunities or continued use of an ineffective method.


Question 2:What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?
 - The P-value is the probability of observing a test statistic at least as extreme as the one calculated from the sample data, assuming the null hypothesis is true.
 Explanation:-   
 Interpretation: In the context of the null hypothesis, a small P-value (typically less than a predetermined significance level, suggests that the observed data is unlikely under the null hypothesis, leading to the rejection of \(H_{0}\). A large P-value suggests the data is consistent with the null hypothesis, leading to the failure to reject \(H_{0}\).


Question 3:Explain the difference between a Z-test and a T-test, including when to use each.
 - Z-test: When to use: Used when the population standard deviation (\(\sigma \)) is known, or when the sample size (\(n\)) is large (typically \(n\ge 30\)), allowing the use of the known population variance or the sample variance as a good estimate. Distribution: Uses the standard normal distribution (Z-distribution).

 T-test: When to use: Used when the population standard deviation (\(\sigma \)) is unknown and must be estimated from the sample standard deviation (\(s\)), and the sample size (\(n\)) is small. Distribution: Uses the t-distribution, which accounts for the added uncertainty of estimating the population standard deviation from a small sample.


Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?
 - Confidence Interval and Margin of Error-
A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain an unknown population parameter (e.g., the true population mean) with a specified level of confidence (e.g., 95%).
Influence of the Margin of Error (ME)
The CI is calculated as:

The Margin of Error (ME) directly influences the CI's width and its interpretation of precision:
 * Width: The ME dictates the width of the interval. A larger ME results in a wider CI (less precise estimate), while a smaller ME results in a narrower CI (more precise estimate).
 * Precision/Confidence Trade-off: Increasing the required confidence level (e.g., from 90% to 99%) increases the critical value, which in turn increases the ME and widens the interval. This means you gain greater certainty (confidence) by sacrificing precision (a wider range of values).


Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?
 - The purpose of an ANOVA test is to determine if there are statistically significant differences between the means of two or more independent groups.

The assumptions of an ANOVA (Analysis of Variance) test are:

 Independence: Observations within and between groups must be independent.
Normality: The data in each group should be approximately normally distributed.
Homogeneity of variances: The variance among the different groups should be approximately equal (homoscedasticity).

ANOVA extends hypothesis testing to more than two groups by comparing the variance between the groups to the variance within the groups using the F-statistic. This allows testing a single null hypothesis, which is more efficient than performing multiple pairwise t-tests that would increase the chance of a Type I error.




```python
#Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

import numpy as np
from statsmodels.stats.weightstats import ztest as ztest

data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 105, 109, 109, 109, 110, 112, 112, 113, 114, 115]

# Hypothesized population mean (H0: mu = 100)
pop_mean_hypothesized = 100

# Perform the one-sample z-test
# The function returns the z-statistic and the p-value
z_statistic, p_value = ztest(data, value=pop_mean_hypothesized)

print(f"Sample Mean: {np.mean(data)}")
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpretation (using a standard alpha level of 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nInterpretation: Reject the null hypothesis (H0). There is sufficient evidence to conclude the sample mean is significantly different from the population mean of 100.")
else:
    print("\nInterpretation: Do not reject the null hypothesis (H0). There is not sufficient evidence to conclude the sample mean is significantly different from the population mean of 100.")
```

```python
#Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using NumPy and plot the histogram.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters for the binomial distribution
n_trials = 10
p_success = 0.5
sample_size = 1000

# Simulate the dataset
# Each value represents the number of successes in 10 trials
simulated_data = np.random.binomial(n=n_trials, p=p_success, size=sample_size)

# Print a few sample values to verify
print(f"First 10 simulated values: {simulated_data[:10]}")

# Plot the histogram
plt.figure(figsize=(8, 5))
sns.histplot(simulated_data, bins=range(n_trials + 2), kde=False, color="blue", edgecolor="black")
plt.title(f"Histogram of Binomial Distribution (n={n_trials}, p={p_success})")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
#Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

import numpy as np
import matplotlib.pyplot as plt

# 1. Non-normal population (Exponential)
np.random.seed(42)
population = np.random.exponential(scale=2, size=10000)
sample_means = []
sample_size = 30
num_samples = 1000

# 2. Draw samples and calculate means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# 3. Visualization to show the result is Normal
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(population, bins=50)
plt.title('Non-Normal Population')
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50)
plt.title('Distribution of Sample Means (Normal)')
plt.show()
```

```python
#Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_and_visualize_ci(data, conf=0.95):
    """Calculates and visualizes the confidence interval."""
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data) # Standard Error of the Mean

    # Calculate the CI
    ci_lower, ci_upper = stats.t.interval(
        confidence=conf,
        df=n - 1,
        loc=mean,
        scale=std_err
    )

    print(f"Sample Mean: {mean:.2f}")
    print(f"{conf*100:.0f}% CI: ({ci_lower:.2f}, {ci_upper:.2f})")

    # Visualization
    plt.errorbar(mean, 0, xerr=[[mean - ci_lower], [ci_upper - mean]],
                 fmt='o', capsize=5, label='CI', color='blue')
    plt.axvline(mean, color='red', linestyle='--', label='Mean')
    plt.title('Confidence Interval')
    plt.yticks([])
    plt.legend()
    plt.show()
    return ci_lower, ci_upper

# Example usage
data = [12, 15, 14, 10, 13, 17, 14, 15, 16, 14]
calculate_and_visualize_ci(data)
```

```python
#Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Example: Testing if a six-sided die is fair
observed = np.array([25, 15, 18, 22, 21, 19])
# Expected for a fair die (120 rolls total): 120 / 6 = 20 each
expected = np.array([20, 20, 20, 20, 20, 20])

# Perform the Chi-square test
chi2, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2:.3f}")
print(f"P-value: {p_value:.3f}")

# --- Explain the Outcome ---
alpha = 0.05
print("\n--- Outcome Explanation ---")
if p_value < alpha:
    print("*Conclusion: Reject Null Hypothesis.* The observed distribution is significantly different from the expected (e.g., the die is unfair).")
else:
    print("*Conclusion: Fail to Reject Null Hypothesis.* There is no significant difference (e.g., the die is fair).")

# Visualization
categories = [f'Side {i+1}' for i in range(len(observed))]
x = np.arange(len(categories))
plt.bar(x - 0.2, observed, 0.4, label='Observed')
plt.bar(x + 0.2, expected, 0.4, label='Expected')
plt.xticks(x, categories)
plt.title('Observed vs. Expected Frequencies')
plt.legend()
plt.show()
```
